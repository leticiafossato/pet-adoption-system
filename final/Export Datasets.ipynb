# Adoption System

## Dataset Manipulation

### Import Libraries

```python
import pandas as pd
```

### Import Kaggle Datasets

```python
#Import train dataset
df = pd.read_csv('../train/train.csv')

#Import test dataset
df_ong = pd.read_csv('../test/test.csv')

#Display all columns
pd.set_option('display.max_columns', None)

#Check the first lines
df.head()
```

### Create target 'NonAdopted'

#### Train

```python
#Creating a new column called 'NonAdopted'

# 'NonAdopted' = 1 (não adotado), 'NonAdopted'=0 (adotado)
indices = df[df['AdoptionSpeed']==4].index
df['NonAdopted'] = 0
df.iloc[indices,df.columns.get_loc("NonAdopted")] = 1
df
```

### Create variable Fee

#### Train

```python
#Creating a new column called 'Fee_y'

# 'Fee_y' = 0 (gratuito), 'Fee_y' = 1 (não gratuito)
indices = df[df['Fee']!=0].index
df['Fee_y'] = 0
df.iloc[indices,df.columns.get_loc('Fee_y')] = 1
df[['Fee','Fee_y']]
```

#### Test (ONG)

```python
#Creating a new column called 'Fee_y'

# 'Fee_y' = 0 (gratuito), 'Fee_y' = 1 (não gratuito)
indices = df_ong[df_ong['Fee']!=0].index
df_ong['Fee_y'] = 0
df_ong.iloc[indices,df_ong.columns.get_loc('Fee_y')] = 1
df_ong[['Fee_y']]
```

### Create  Photo column

#### Test (ONG)

```python
lista=[]
for i,j in zip(df_ong['PetID'],df_ong['PhotoAmt']):
    a=(i+'-'+str(int(j)))
    lista.append(a)
df_ong['Photo'] = lista
```

### Replace NaN Names

#### Test (ONG)

```python
df_ong.loc[df_ong['Name'].isna(),'Name'] = 'No Name'
```

### Quantity Analysis

In this analysis, quantity will genarete wrong analysis.
Because the status of NonAdoption could be 1 (non adopted) because the quantity is high.

#### Drop columns test (ONG)

```python
df_ong2 = df_ong.drop(df_ong[(df_ong['Quantity']!=1)].index)
```

```python
df_ong2.head()
```

#### Drop columns train

```python
df2 = df.drop(df[(df['Quantity']!=1)].index)
```
